## 1. Get webpage using *requests*

```python
import requests

req = requests.get('https://www.liaoxuefeng.com/wiki/896043488029600')
```

```python
webpage = req.text
```

## 2. Get specific contents using BeatifulSoup

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(webpage, 'html.parser')
```

### 2.1 Prettify the webpage

```python
print(soup.prettify())
```

### 2.2 Get the first paragraph

You can try to remove "attrs" to see how it works.

```python
paragraph = soup.find_all('p')
```

```python
paragraph
```

```python
paragraph = soup.find('p', attrs={"class":False})
```

```python
paragraph
```

### 2.3 Get all the links in this paragraph which point to other webpages

```python
soup.find_all('a')
```

```python
soup.find_all('a', attrs={"href":True, 'class':"x-wiki-index-item"})
```

```python
soup.find_all('a', attrs={"href":True, 'class':"x-wiki-index-item"})[0].string
```

```python
data = {"title":[], "href":[]}
for link in soup.find_all('a', attrs={"href":True, 'class':"x-wiki-index-item"}):
    data["title"].append(link.string)
    data["href"].append(link["href"])
```

```python
import pandas as pd
df = pd.DataFrame(data)
```

```python
df
```

## 3. Get the contents from all the webpages

```python
webpages = []
head = "https://www.liaoxuefeng.com"
for href in data["href"]:
    link = head + href
    req = requests.get(link)
    webpage = req.text
    webpages.append(webpage)
```

```python
f = open('liaoxuefeng.txt', "w")
f.write(webpages[0])
```
